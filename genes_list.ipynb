```python
### creating/renaming the columns containing the gene names in such way that they are all unified --> e.g. for

# Function to standardize column names for gene lists
def standardize_gene_column(df):
    # For marker gene datasets (with 'x' column)
    if 'x' in df.columns:
        df = df.rename(columns={'x': 'Gene'})
        if 'Unnamed: 0' in df.columns:
            df = df.drop('Unnamed: 0', axis=1)
    
    # For correlation datasets (where genes are in 'Unnamed: 0')
    elif 'Unnamed: 0' in df.columns and 'corr_vector' in df.columns:
        df = df.rename(columns={'Unnamed: 0': 'Gene'})
    
    return df

# Apply to all datasets
# Dataset 1, tp3.5
data1_tp3 = standardize_gene_column(data1_tp3)
d3_1_spapros = standardize_gene_column(d3_1_spapros)
d3_1_corr = standardize_gene_column(d3_1_corr)
d3_1_ciara = standardize_gene_column(d3_1_ciara)


# Dataset 1, tp4.5
data1_tp4 = standardize_gene_column(data1_tp4)
d4_1_spapros = standardize_gene_column(d4_1_spapros)
d4_1_corr = standardize_gene_column(d4_1_corr)
d4_1_ciara = standardize_gene_column(d4_1_ciara)


# Dataset 2, tp3.5
data2_tp3 = standardize_gene_column(data2_tp3)
d3_2_spapros = standardize_gene_column(d3_2_spapros)
d3_2_corr = standardize_gene_column(d3_2_corr)
d3_2_ciara = standardize_gene_column(d3_2_ciara)

# Dataset 2, tp4.5
data2_tp4 = standardize_gene_column(data2_tp4)
d4_2_spapros = standardize_gene_column(d4_2_spapros)
d4_2_corr = standardize_gene_column(d4_2_corr)
d4_2_ciara = standardize_gene_column(d4_2_ciara)

# Verify the changes for one dataset from each type
print("Marker gene dataset columns:", data1_tp3.columns)
print("Spapros dataset columns:", d3_1_spapros.columns)
print("Correlation dataset columns:", d3_1_corr.columns)
print("Ciara dataset columns:", d3_1_ciara.columns)
```

```python
# First add source labels as before
# Dataset 1, tp3.5
data1_tp3['source'] = 'marker'
d3_1_spapros['source'] = 'spapros'
d3_1_corr['source'] = 'lefty_corr'
d3_1_ciara['source'] = 'potential_markers'

# Dataset 1, tp4.5
data1_tp4['source'] = 'marker'
d4_1_spapros['source'] = 'spapros'
d4_1_corr['source'] = 'lefty_corr'
d4_1_ciara['source'] = 'potential_markers'


# Dataset 2, tp3.5
data2_tp3['source'] = 'marker'
d3_2_spapros['source'] = 'spapros'
d3_2_corr['source'] = 'lefty_corr'
d3_2_ciara['source'] = 'potential_markers'


# Dataset 2, tp4.5
data2_tp4['source'] = 'marker'
d4_2_spapros['source'] = 'spapros'
d4_2_corr['source'] = 'lefty_corr'
d4_2_ciara['source'] = 'potential_markers'



def create_gene_summary(datasets):
    # Initialize empty DataFrame with all genes
    all_genes = pd.concat([df['Gene'] for df in datasets]).unique()
    final_df = pd.DataFrame({'Gene': all_genes})
    
    # Initialize columns with False
    final_df['marker_gene'] = False
    final_df['spapros'] = False
    final_df['lefty1_correlation'] = False
    final_df['potential_markers'] = False
    
    # Fill in True values based on source
    for df in datasets:
        genes = df['Gene'].unique()
        if df['source'].iloc[0] == 'marker':
            final_df.loc[final_df['Gene'].isin(genes), 'marker_gene'] = True
        elif df['source'].iloc[0] == 'spapros':
            final_df.loc[final_df['Gene'].isin(genes), 'spapros'] = True
        elif df['source'].iloc[0] == 'lefty_corr':
            final_df.loc[final_df['Gene'].isin(genes), 'lefty1_correlation'] = True
        elif df['source'].iloc[0] == 'potential_markers':
            final_df.loc[final_df['Gene'].isin(genes), 'potential_markers'] = True

    
    # Add total count of Trues for sorting
    final_df['total_hits'] = final_df[['marker_gene', 'spapros', 'lefty1_correlation', 'potential_markers']].sum(axis=1)
    
    # Sort by total hits (descending) and then by gene name
    final_df = final_df.sort_values(['total_hits', 'Gene'], ascending=[False, True])
    
    # Drop the total_hits column as it was just for sorting
    final_df = final_df.drop('total_hits', axis=1)
    
    return final_df

# Create final datasets
final_data1_tp3 = create_gene_summary([data1_tp3, d3_1_spapros, d3_1_corr, d3_1_ciara])
final_data1_tp4 = create_gene_summary([data1_tp4, d4_1_spapros, d4_1_corr, d4_1_ciara])
final_data2_tp3 = create_gene_summary([data2_tp3, d3_2_spapros, d3_2_corr, d3_2_ciara])
final_data2_tp4 = create_gene_summary([data2_tp4, d4_2_spapros, d4_2_corr, d4_2_ciara])

# Example output
print("Dataset 1, tp3.5 sample (sorted by number of hits):")
print(final_data1_tp3.head(10))
print('shape of dataset 1 tp3.5:', final_data1_tp3.shape)

final_data2_tp4.to_csv('dataset2_tp4.5_genes_threshold.csv', index=False) # Save to CSV if needed
final_data1_tp3.to_csv('dataset1_tp3.5_genes_threshold.csv', index=False)
final_data1_tp4.to_csv('dataset1_tp4.5_genes_threshold.csv', index=False)
final_data2_tp3.to_csv('dataset2_tp3.5_genes_nothreshold.csv', index=False)
```

```python

# Creating an empty dictionary for lookup
gene_celltype_lookup = {}

# Loop through each cell type column in `known_genes`
for cell_type in known_genes.columns[1:]:  # skip 'Cell Types' column
    genes_in_cell_type = known_genes[cell_type].dropna().tolist()
    for gene in genes_in_cell_type:
        gene_celltype_lookup[gene.upper()] = cell_type  # map gene to cell type

# Step 2: Define a function to find cell type for each gene
def add_known_gene_of_celltype_column(dataset, gene_celltype_lookup):
    # Initialize an empty list to store the cell types for each gene
    celltype_annotations = []
    # Iterate over each gene in the dataset
    for gene in dataset['Gene']:
        # Look up the gene in the dictionary, default to None if not found
        celltype = gene_celltype_lookup.get(gene.upper(), None)
        celltype_annotations.append(celltype)
    # Add a new column to the dataset
    dataset['known_gene_of_celltype'] = celltype_annotations
    return dataset

# Step 3: Apply the function to each of the four datasets
# Assuming each dataset has a 'gene' column listing the genes
# Replace `dataset1_tp3.5`, `dataset1_tp4.5`, etc., with your actual dataset variables
dataset1_tp3_genes = add_known_gene_of_celltype_column(final_data1_tp3, gene_celltype_lookup)
dataset1_tp4_genes = add_known_gene_of_celltype_column(final_data1_tp4, gene_celltype_lookup)
dataset2_tp3_genes = add_known_gene_of_celltype_column(final_data2_tp3, gene_celltype_lookup)
dataset2_tp4_genes = add_known_gene_of_celltype_column(final_data2_tp4, gene_celltype_lookup)

# Now each dataset will have a 'known_gene_of_celltype' column with matched cell types
```
